# Preprocessing and Modelling

To build the corpus from the 'verified_reviews' we perform the following - <br>
1. Replace any non alphabet characters with a space
2. Covert to lower case and split into words
3. Iterate over the individual words and if it is not a stopword then add the stemmed form of the word to the corpus

```python
corpus = []
stemmer = PorterStemmer()
for i in range(0, data.shape[0]):
  review = re.sub('[^a-zA-Z]', ' ', data.iloc[i]['verified_reviews'])
  review = review.lower().split()
  review = [stemmer.stem(word) for word in review if not word in STOPWORDS]
  review = ' '.join(review)
  corpus.append(review)
```

Using Count Vectorizer to create bag of words

```python
cv = CountVectorizer(max_features = 2500)

#Storing independent and dependent variables in X and y
X = cv.fit_transform(corpus).toarray()
y = data['feedback'].values
```

```python
#Saving the Count Vectorizer
pickle.dump(cv, open('countVectorizer.pkl', 'wb'))
```

Checking the shape of X and y

```python
print(f"X shape: {X.shape}")
print(f"y shape: {y.shape}")
```

Splitting data into train and test set with 30% data with testing.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 15)

print(f"X train: {X_train.shape}")
print(f"y train: {y_train.shape}")
print(f"X test: {X_test.shape}")
print(f"y test: {y_test.shape}")
```

```python
print(f"X train max value: {X_train.max()}")
print(f"X test max value: {X_test.max()}")
```

We'll scale X_train and X_test so that all values are between 0 and 1.

```python
scaler = MinMaxScaler()

X_train_scl = scaler.fit_transform(X_train)
X_test_scl = scaler.transform(X_test)
```

```python
import re
from io import BytesIO
import pandas as pd
import pickle
import matplotlib.pyplot as plt
import base64
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

# Load models and tools
predictor = pickle.load(open(r"model_xgb.pkl", "rb"))
scaler = pickle.load(open(r"scaler.pkl", "rb"))
cv = pickle.load(open(r"countVectorizer.pkl", "rb"))

STOPWORDS = set(stopwords.words("english"))

# Single string prediction function
def single_prediction(predictor, scaler, cv, text_input):
    corpus = []
    stemmer = PorterStemmer()
    review = re.sub("[^a-zA-Z]", " ", text_input)
    review = review.lower().split()
    review = [stemmer.stem(word) for word in review if word not in STOPWORDS]
    review = " ".join(review)
    corpus.append(review)
    X_prediction = cv.transform(corpus).toarray()
    X_prediction_scl = scaler.transform(X_prediction)
    y_predictions = predictor.predict_proba(X_prediction_scl)
    y_predictions = y_predictions.argmax(axis=1)[0]

    return "Positive" if y_predictions == 1 else "Negative"

# Bulk prediction function
def bulk_prediction(predictor, scaler, cv, data):
    corpus = []
    stemmer = PorterStemmer()
    for i in range(0, data.shape[0]):
        review = re.sub("[^a-zA-Z]", " ", data.iloc[i]["Sentence"])
        review = review.lower().split()
        review = [stemmer.stem(word) for word in review if word not in STOPWORDS]
        review = " ".join(review)
        corpus.append(review)

    X_prediction = cv.transform(corpus).toarray()
    X_prediction_scl = scaler.transform(X_prediction)
    y_predictions = predictor.predict_proba(X_prediction_scl)
    y_predictions = y_predictions.argmax(axis=1)
    y_predictions = list(map(sentiment_mapping, y_predictions))

    data["Predicted sentiment"] = y_predictions
    display(data)

    # Generate distribution graph
    get_distribution_graph(data)

# Sentiment mapping function
def sentiment_mapping(x):
    return "Positive" if x == 1 else "Negative"

# Function to generate and display the sentiment distribution graph
def get_distribution_graph(data):
    fig = plt.figure(figsize=(5, 5))
    colors = ("green", "red")
    wp = {"linewidth": 1, "edgecolor": "black"}
    tags = data["Predicted sentiment"].value_counts()
    explode = (0.01, 0.01)

    tags.plot(
        kind="pie",
        autopct="%1.1f%%",
        shadow=True,
        colors=colors,
        startangle=90,
        wedgeprops=wp,
        explode=explode,
        title="Sentiment Distribution",
        xlabel="",
        ylabel="",
    )

    plt.show()

# Example usage:
# For single prediction:
text_input = "I love this product! It's fantastic."
predicted_sentiment = single_prediction(predictor, scaler, cv, text_input)
print(f"Predicted Sentiment: {predicted_sentiment}")

# For bulk prediction:
# Assuming 'data' is a pandas DataFrame with a column 'Sentence'
# data = pd.read_csv('SentimentBulk.csv')
# bulk_prediction(predictor, scaler, cv, data)
```
